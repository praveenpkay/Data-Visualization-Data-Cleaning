```python
# Pre-processing

# Bookin ID must be distinct
df.drop_duplicates(subset ="BookingID",
                     keep = False, inplace = True)
# 953 values Gps provider missing

# Expected ETA is 1899 in row 6868, 6264
df.drop(df.index[[6868,6264]], inplace=True)

#single cloumn for Delay/Ontime
df['ontime/delay']=df.ontime.replace({np.NaN, 'G'}, {0, 1})
```

```python
df['ontime/delay'].value_counts()
```

```python
df.head()
```

```python
df.shape
```

```python
df['vehicle_states'] = df.vehicle_no.astype(str).apply(lambda x:x.upper())
df['vehicle_states'] = df['vehicle_states'].str.extract(r'(^[a-zA-Z]{2})')
df['vehicle_states'].value_counts()
```

```python
df.head()
```

```python
df.drop(labels=['ontime',	'delay','DestinationLocation'],axis=1, inplace=True)
```

```python
df['Origin_Location_']= df.Origin_Location.astype(str).apply(lambda x:x.split(',')[-1]).apply(lambda x : x.upper().strip())
df['Origin_Location_'].unique()
# Handling Values with India as Origin Location
t_df = df[df['Origin_Location_'] == 'INDIA']
df = df.drop(t_df.index)
t_df['Origin_Location_']= t_df.Origin_Location.astype(str).apply(lambda x:x.split(',')[-2]).apply(lambda x : x.upper().strip())
t_df['Origin_Location_'].unique()
df = pd.concat([df, t_df], axis=0)
df['Origin_Location_'] = df['Origin_Location_'].str.strip().replace('PUDUCHERRY','PONDICHERRY')
df['Origin_Location_'] = df['Origin_Location_'].str.strip().replace('\d+', '', regex=True)
df['Origin_Location_'].unique()
```

```python
df['Destination_Location_']= df.Destination_Location.astype(str).apply(lambda x:x.split(',')[-1]).apply(lambda x : x.upper().strip())
df['Destination_Location_'].unique()
# Handling Values with India as Origin Location
t_df = df[df['Destination_Location_'] == 'INDIA']
df = df.drop(t_df.index)
t_df['Destination_Location_']= t_df.Destination_Location.astype(str).apply(lambda x:x.split(',')[-2]).apply(lambda x : x.upper().strip())
t_df['Destination_Location_'].unique()
df = pd.concat([df, t_df], axis=0)
df['Destination_Location_'] = df['Destination_Location_'].str.strip().replace('PUDUCHERRY','PONDICHERRY')
df['Destination_Location_'] = df['Destination_Location_'].str.strip().replace('\d+', '', regex=True)
df['Destination_Location_'].unique()
```

```python
df.shape
```

```python
#let's check the percentage of null values in each feature
for col in df.columns:
    if df[col].isna().sum()>0:
        print(col, df[col].isna().mean().round(4)*100)
```

```python
df.shape
```

```python
df['Driver_Name']=df['Driver_Name'].fillna('Unknown')
df['vehicleType']=df['vehicleType'].fillna('Unknown')
df['GpsProvider']=df['GpsProvider'].fillna('Unknown')
df.drop(labels=['Destination_Location','Origin_Location'],axis=1, inplace=True)
```
